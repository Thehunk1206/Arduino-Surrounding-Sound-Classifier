# FFT Audio Data EDA

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
data = np.load("Dataset/whistle.npy")
```

```python
def plot_all_fft_audio():  
    for i in range(10):      
        fig = plt.figure()
        ax = plt.axes()
        #plt.subplot(2,5, i+1)
        plt.plot(data[i].reshape(-1,1))
        plt.show()
plot_all_fft_audio()
```
